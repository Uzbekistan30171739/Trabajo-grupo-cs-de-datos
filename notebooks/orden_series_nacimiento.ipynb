```python
import pandas as pd
```

```python
df = pd.read_csv("Serie_Nacimientos_1992_2022.csv", encoding= 'latin1', sep= ';')
```

```python
df = df.drop(columns=['GLOSA_REGION_RESIDENCIA'] )
```

```python
regiones_chile = {
    15: 'ARICA Y PARINACOTA',
    1: 'TARAPACA',
    2: 'ANTOFAGASTA',
    3: 'ATACAMA',
    4: 'COQUIMBO',
    5: 'VALPARAISO',
    13: 'METROPOLITANA',
    6: 'OHIGGINS',
    7: 'MAULE',
    16: 'NUBLE',
    8: 'BIOBIO',
    9: 'LA ARAUCANIA',
    14: 'LOS RIOS',
    10: 'LOS LAGOS',
    11: 'AYSEN',
    12: 'MAGALLANES',
    99: 'NO IDENTIFICADO'
}
```

```python
df['REGION_RESIDENCIA'] = df['REGION_RESIDENCIA'].map(regiones_chile)
```

```python
df
```

```python
df.columns = ["Mes", "Anio", "Sexo", "Rango_edad_M", "Educacion_M", "Rango_edad_P", "Educacion_P", "Region"]
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df['Rango_edad_M'] = df['Rango_edad_M'].str.replace(' AÑOS', '')
```

```python
df['Rango_edad_P'] = df['Rango_edad_P'].str.replace(' AÑOS', '')
```
