```python
import pandas as pd
import numpy as np
import os
```

```python
columnas_a_leer = [
    'periodo',
    'comuna',
    'n_personas_0_40_comuna',
    'n_personas_41_50_comuna',
    'n_personas_51_60_comuna',
    'n_personas_61_70_comuna',
    'n_personas_71_80_comuna',
    'n_personas_81_90_comuna',
    'n_personas_91_100_comuna',
    'total_comuna'
]
```

```python
df = pd.read_csv("DATOS_CONSOLIDADOS_RSH_limpio.csv", usecols=columnas_a_leer)
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python

for i in range(3, 10):
    columna = df.columns[i]

    df[columna] = pd.to_numeric(df[columna], errors='coerce')

    df[columna] = df[columna].fillna(5)

    df[columna] = df[columna].astype('Int64')
```
